# MUMBAI CITY 

```python
#### for mumbai city
#import time
#start = time.time()
"the code you want to test stays here"

import scrapy
from scrapy.crawler import CrawlerProcess

import pandas as pd
from urllib.parse import urljoin

## creating list in which the data can be stored
status1 = []
sqt2 = []
price2 =[]
pricecolumn2 =[]
bhk2 = []
apartment_name2 =[]
possession_date =[]
location2 = []
size2 = []
city2 =[]
latitude2 = []
longitude2 =[]
seller_type2=[]

for page in range(1, 5000): #no. of pages gives as required
    url = f"https://www.makaan.com/mumbai-residential-property/buy-property-in-mumbai-city?_=1681564791453&page={page}" # page 1
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    for property in soup.find_all('li', class_='cardholder'):  #main card which holds all information about the price, sqft, building_name, size, cityname and many more
        
        status = property.find('td', class_='val')# status of the building
        
        
        sqt = property.find('td',class_="lbl rate") # sqft
        
        price = property.find('td',class_= 'price').find('span', class_ = 'val')
        #print(price.text.strip())
        pricetag = property.find('span', class_ = 'unit')
        #print(pricetag.text.strip())
        bhk = property.find('strong')
        #print(BHK.text.strip())
        building_name = property.find('span', class_='project-wrap')
        #print(building_name.text.strip())
        possession1 = property.find('li',class_ = 'keypoint')
        #print(possession.text.strip())
        location = property.find('span', class_ ='locName')
        #print(location.text.strip())
        size = property.find('td' ,class_="size")
        #print(size.text.strip())
        cityname = property.find('span',class_ = 'locName')
        #print(cityname.text.strip())
        latitude = property.find('meta',{'itemprop' :"latitude"})
        #print(location.get('content'))
        longitude = property.find('meta',{'itemprop' :"longitude"})
        
        seller = property.find('span', class_ = "seller-type")
        #print(seller.text.strip())
   
    
    # appending all values in our empty list
        
        if status is not None:
            status1.append(status.string)
        if sqt is not None:
            sqt2.append(sqt.string)
        if price is not None:
            price2.append(price.text.strip())
        if pricetag is not None:
            pricecolumn2.append(pricetag.text.strip())
        if bhk is not None:
            bhk2.append(bhk.text.strip())
        if building_name is not None:
            apartment_name2.append(building_name.text.strip())
        elif building_name is None:
            apartment_name2.append("NA") ### there were some missing values 
        if possession1 is not None:
            possession_date.append(possession1.text.strip())
        elif possession1 is None:
            possession_date.append("NA")
        if location is not None:
            location2.append(location.text.strip())
        elif location is None:
            location2.append("NA")
        if size is not None:
            size2.append(size.text.strip())
        if cityname is not None:
            city2.append(cityname.text.strip())
        elif cityname is None:
            city2.append("NA")
        if latitude is not None:
            latitude2.append(latitude.get('content'))
        if longitude is not None:
            longitude2.append(longitude.get('content'))
        if seller is not None:
            seller_type2.append(seller.text.strip())

# creating a dataframe and all the len of the values must be same so we have given the above condition for None = "NA"

df = { "status" : status1, "sqtr" : sqt2, "prices" : price2 , "price_tag" : pricecolumn2 , "bhk": bhk2, "Apartment": apartment_name2, "possesion": possession_date,
      "size": size2 , "latitude" : latitude2 , "longitude" : longitude2 , "seller_type" :seller_type2 , "city" : city2
      
      }
 
# save in our local system
mumbai_city = pd.DataFrame(df)      
mumbai_city.to_csv('/Users/raw/Desktop/mumbai_city.csv')        

            
#end = time.time()
#print(end - start)

# it took around 186min to scrap
```
